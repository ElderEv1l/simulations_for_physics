```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
c = 2
U = 1
L = 30
P1 = 7
P2 = 12

def get_V(x):
    V = np.zeros_like(x)

    for i, xi in enumerate(x):
        n = np.floor(xi / c)
        if n * c < xi < n * c + a:
            V[i] = 0
        elif n * c + a < xi < (n + 1) * c:
            V[i] = U

    return V

def get_Kronig_Penney(P):
    x = np.linspace(-L, L, 10000)
    plus_one = np.full(shape=len(x), fill_value=1, dtype=np.int64)
    minus_one = np.full(shape=len(x), fill_value=-1, dtype=np.int64)
    y = P * np.sin(x) / x + np.cos(x)
    plt.plot(x, y, label=r'$P = ${}'.format(P))
    plt.plot(x, plus_one, 'k--')
    plt.plot(x, minus_one, 'k--')
    plt.fill_between(x, plus_one, -1, where=(np.abs(y) < 1), color='g', alpha = 0.5)
    plt.xlabel("a * alpha")
    plt.ylabel("P * sin(a * alpha) / (a * alplha) + cos(a * alpha)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
print("Рассмотрим для случая свободного электрона")
get_Kronig_Penney(0)
```

```python
print("Когда электрон заперт внутри одной потенциальной ямы")
get_Kronig_Penney(9999999999999999)
```

```python
print("Также рассмотрим промежуточные случаи")
get_Kronig_Penney(P1)
get_Kronig_Penney(P2)
```
